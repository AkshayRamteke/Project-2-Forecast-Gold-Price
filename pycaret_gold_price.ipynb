```python
# creating a 12 months moving average
data["MA12"] = data['Price'].rolling(12).mean()
data
```

```python
# Adding series of numbers
import numpy as np
data['seq_num'] = np.arange(1,len(data)+1)
# Dropping unnecessary columns from the dataset

data2 = data[['seq_num','month','year','Price']]
data2.head()
```

```python
# Splitting the data into train and test
train = data2[data2['year']<2008]
test = data2[data2['year']>=2008]
train.shape,test.shape
```
